```python
# 문자열에서 쉼표 제거후 정수형으로 변환
df = df.replace(",","",regex=True).astype(int)
df
```

```python
# 컬럼 순서 변경, 주요 국가만 추출
df = df[['OECD평균', '한국', '미국', '호주','일본', '프랑스', '독일', '영국', '이탈리아']]
df.head()
```

```python
plt.figure(figsize=(12, 8))
plt.plot(df.index,df["한국"], marker='o', label="한국")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 선택한 국가들의 연도별 1인당 전력 소비량을 꺾은선 그래프로 그리는 함수

def plot_energy_consumption(data, countries):

    # 연도를 x축으로 설정
    years = df.index

    # 국가별로 꺾은선 그래프 그리기
    plt.figure(figsize=(12, 8))
    for country in countries:
        plt.plot(years, df[country], marker='o', label=country)

    plt.title('주요 국가 1인당 전력 소비량 (2001-2019)')
    plt.xlabel('년도')
    plt.ylabel('1인당 전력 소비량 (kWh)')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)
    plt.show()

# 함수 호출
plot_energy_consumption(df, ['OECD평균','한국', '미국'])
```
